```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
data = pd.DataFrame()
for file in os.listdir():
    if file.split('.')[-1] == 'parquet':
        print(f'Loading: {file}')
        new_data = pd.read_parquet(file)
        data = pd.concat([data, new_data])
data
```

```python
data['label_combination'] = data[['baseline_drift', 'static_noise', 'burst_noise', 'electrodes_problems', 'clean']].apply(
    lambda row: ','.join([col for col, val in zip(row.index, row) if val == 1.0]),
    axis=1
)

# Get the distribution of these label combinations
label_distribution = data['label_combination'].value_counts()
print(label_distribution)
```
